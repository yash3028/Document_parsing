```python
from datasets import load_dataset

dataset = load_dataset(
    "jordanparker6/publaynet",
    split="train",
    streaming=True
)
```

```python
from itertools import islice

subset = list(islice(dataset, 2000))
```

```python
subset[0]
```

```python
import os

os.makedirs("dataset/images/train", exist_ok=True)
os.makedirs("dataset/labels/train", exist_ok=True)
```

```python
category_map = {
    1: 0,   # text
    2: 1,   # title
    3: 2,   # list
    4: 3,   # table
    5: 4    # figure
}
```

```python
category_map[1]
```

```python
for i, sample in enumerate(subset):

    # Save image
    sample["image"].save(f"dataset/images/train/{i}.jpg")

    img_width, img_height = sample["image"].size

    # Create label file
    with open(f"dataset/labels/train/{i}.txt", "w") as f:

        for ann in sample["annotations"]:

            x, y, w, h = ann["bbox"]

            # Convert COCO -> YOLO
            x_center = (x + w/2) / img_width
            y_center = (y + h/2) / img_height
            w = w / img_width
            h = h / img_height

            class_id = category_map[ann["category_id"]]

            f.write(
                f"{class_id} "
                f"{x_center:.6f} "
                f"{y_center:.6f} "
                f"{w:.6f} "
                f"{h:.6f}\n"
            )
```

```python
import os
import shutil
import random

# image names without extension
files = [f.split(".")[0] for f in os.listdir("dataset/images/train")]

random.seed(42)
random.shuffle(files)

train_files = files[:1500]
val_files = files[1500:1800]
test_files = files[1800:]

# create folders
for folder in [
    "dataset/images/val",
    "dataset/images/test",
    "dataset/labels/val",
    "dataset/labels/test"
]:
    os.makedirs(folder, exist_ok=True)

# move validation files
for f in val_files:
    shutil.move(f"dataset/images/train/{f}.jpg",
                f"dataset/images/val/{f}.jpg")
    shutil.move(f"dataset/labels/train/{f}.txt",
                f"dataset/labels/val/{f}.txt")

# move test files
for f in test_files:
    shutil.move(f"dataset/images/train/{f}.jpg",
                f"dataset/images/test/{f}.jpg")
    shutil.move(f"dataset/labels/train/{f}.txt",
                f"dataset/labels/test/{f}.txt")
```
